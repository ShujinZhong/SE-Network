```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Plot evaluation for cifar10

```python
res_cifar_acc = pd.read_csv('../../Data/Resnet50-SE/resnet_cifar10_accuracy.csv', names=['Resnet_50_train'])
res_cifar_val_acc = pd.read_csv('../../Data/Resnet50-SE/resnet_cifar10_val_accuracy.csv', names=['Resnet_50_val'])
se_res_cifar_acc = pd.read_csv('../../Data/Resnet50-SE/SE_resnet_cifar10_accuracy.csv', names=['SE_Resnet_50_train'])
se_res_cifar_val_acc = pd.read_csv('../../Data/Resnet50-SE/SE_resnet_cifar10_val_accuracy.csv', names=['SE_Resnet_50_val'])
```

```python
acc_df = res_cifar_acc.merge(res_cifar_val_acc, left_index=True, right_index=True)
acc_df = acc_df.merge(se_res_cifar_acc, left_index=True, right_index=True)
acc_df = acc_df.merge(se_res_cifar_val_acc, left_index=True, right_index=True)
acc_df.head()
```

```python
err_df = acc_df.apply(lambda x: 1 - x)
err_df.head()
```

```python
plt.figure(dpi=100)
ax = plt.subplot(111)
err_df.plot(kind='line', ax=ax, style=[':', '-', ':', '-'], color=['blue', 'blue', 'red', 'red'])
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Error Comparison for Cifar10')
```
