```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
RN110 = pd.read_csv('ResNet110.csv')
RN110_SE = pd.read_csv('SEResNet110.csv')
```

```python
RN110['error'] = 1 - RN110['accuracy']
RN110['val_error'] = 1 - RN110['val_accuracy']
```

```python
RN110['error'].min()
```

```python
RN110['val_error'].min()
```

```python
RN110_SE['error'] = 1 - RN110_SE['accuracy']
RN110_SE['val_error'] = 1 - RN110_SE['val_accuracy']
```

```python
RN110_SE['error'].min()
```

```python
RN110_SE['val_error'].min()
```

```python
plt.figure(figsize=(10,6))
plt.plot(RN110['epoch'], RN110['error'], color = 'blue', linestyle='dashed', label = 'RN_Train')
plt.plot(RN110['epoch'], RN110['val_error'], color = 'blue', label = 'RN_Test')
plt.plot(RN110_SE['epoch'], RN110_SE['error'], color = 'red', linestyle='dashed', label = 'RN-SE_Train')
plt.plot(RN110_SE['epoch'], RN110_SE['val_error'], color = 'red', label = 'RN_SE_Test')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('ResNet110 w/wo SE for cifar10')
plt.legend()
```
